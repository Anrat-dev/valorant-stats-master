```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
path = "data/complete_file_weapons.csv"
df = pd.read_csv(path)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Weapon Name', y='Kills Per Match', data=df)
plt.title('Weapon Name vs Kills Per Match')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Weapon Name', hue='Label', data=df)
plt.title('Weapon Name vs Label')
plt.show()
```

```python
df['Average Shot'] = df[['Headshot', 'Bodyshot', 'Legshot']].mean(axis=1)
plt.figure(figsize=(10,6))
sns.kdeplot(data=df, x='Average Shot', y='Kills Per Match', fill=True)
plt.title('Average Shot vs Kills Per Match')
plt.show()
```
