```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
path = "data/complete_file_agents.csv"
df = pd.read_csv(path)
```

```python
#combatscore (ACS) VS label

#the grpah shows that a higher combat score directly correlates to having a higher winrate as showcased in the box plot
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Label', y='ACS')
plt.title('ACS vs Winrate')
plt.show()
```

```python
#KD vs lABEL
#The graph shows the Kill-to-Death ratio of an agent as compared to the winrate, The higher the KD the higher the winrate of an agent in this case
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='KD', y='Label')
plt.title('KD vs winrate')
plt.show()
```

```python
#aqent to label
#this is an interesting grpah as it shows that the winrate is  largely consistent  across all agents. Shows that there is no particular agent that provides a higher winrate when chosen
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Agent Name', y='Label')
plt.title('Agent Name vs Winrate ')
plt.xticks(rotation=90)  
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Agent Name', y='Kills')
plt.title('Kill Rate vs Agent Name')
plt.xticks(rotation=90)  # This will rotate the x-axis labels for better visibility
plt.show()
```
